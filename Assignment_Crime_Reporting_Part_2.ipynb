```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
```

```python
df = pd.read_csv('C:\Program Files\Python36\Crime_Sample.txt')
```

```python
df['Date'] = pd.to_datetime(df.Date)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Day']  = df.Date.dt.day
```

```python
df['Weekday'] = df.Date.dt.weekday
```

```python
booleans = []
for day in df.Weekday:
    if day > 4:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans1 = []
for day in df.Weekday:
    if day <= 4:
        booleans1.append(True)
    else:
        booleans1.append(False)
```

```python
is_weekend = pd.Series(booleans)
```

```python
is_weekday = pd.Series(booleans1)
```

```python
is_weekend.head()
```

```python
df[is_weekend]
```

```python
crime_weekend = df[is_weekend].groupby('District')['District'].count()
```

```python
crime_weekday = df[is_weekday].groupby('District')['District'].count()
```

```python
crime_weekend
```

```python
crime_weekday
```

```python
print("District wise crimes during weekend")
print(crime_weekend)
print("\n")
print("District wise crimes during weekdays")
print(crime_weekday)
```

```python
# Total crimes during weekends across districts
crime_weekend.sum()
```

```python
# Total crimes during weekdays across districts
crime_weekday.sum()
```

```python
# Maximum crimes in a single district during weekend and weekdays
print(crime_weekend.max())
print(crime_weekday.max())
```
