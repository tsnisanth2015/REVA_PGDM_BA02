```python
import pandas as pd
import numpy as np
from sklearn import datasets
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.show()
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
iris.groupby('species').size()
```

```python
iris.describe()
```

```python
np.round(iris.mean(),2)
```

```python
np.round(iris.median(),2)
```

```python
pd.crosstab(index=iris['species'],columns="Frequency")
```

```python
correlation = iris.corr()
correlation
```

```python
sns.heatmap(correlation,cmap="RdYlGn",annot=True)
plt.show();
sns.heatmap(correlation,cmap="Blues",annot=True)
plt.show();
```

```python
#Compute Percentile
np.round(iris.quantile((np.arange(0,1,.05))),2)
```

```python
#Pairplot
sns.pairplot(iris,hue="species",palette="RdYlGn",size=5)
plt.show();
```

```python
sns.pairplot(iris,hue="species",palette="copper",size=5,x_vars=["sepal_width","sepal_length"],y_vars=["petal_width","petal_length"])
plt.show();
```

```python
sns.pairplot(iris,kind="reg")
plt.show();
```

```python
sns.pairplot(iris,size=3,vars=["sepal_width","sepal_length"])
plt.show();
```

```python
plt.figure(figsize=(15,10))
plt.subplots_adjust(hspace=0.2,wspace=0.2)
plt.subplot(2,2,1)
plt.title("Sepal Length")
plt.plot(iris.sepal_length,'r')
plt.subplot(2,2,2)
plt.title("Sepal Width")
plt.plot(iris.sepal_width,'b')
plt.subplot(2,2,3)
plt.title("Petal Length")
plt.plot(iris.petal_length,'g')
plt.subplot(2,2,4)
plt.title("Petal Width")
plt.plot(iris.petal_width,'y')
plt.show();
```

```python
#Histogram and Kernel Density estimator
plt.figure(figsize=(15,10))
sns.distplot(iris.sepal_width)
plt.show();
```

```python
plt.figure(figsize=(15,10))
sns.distplot(iris.petal_width)
plt.show();
```

```python
#Box Plot
plt.figure(figsize=(15,10))
sns.boxplot(data=iris,orient="v",palette="Set2")
plt.show();
```

```python
#Box Plot
plt.figure(figsize=(15,10))
sns.boxplot(x='species',y='sepal_length',data=iris,linewidth=2)
plt.show();
```

```python
#Heatmap
flight_data = sns.load_dataset("flights")
flight_data.tail()
flight_data.shape
```

```python
flight_data = flight_data.pivot('month','year','passengers')
flight_data
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(flight_data,annot=True,fmt="d")
plt.show();
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(flight_data,cmap="YlGnBu")
plt.show();
```
