```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://www.espn.com/nba/statistics/player/_/stat/assists/sort/avgAssists/year/2018/seasontype/2'
```

```python
headers = {'User-Agent': 'Mozilla/5.0'}
```

```python
response = requests.get(url, headers = headers)
```

```python
response.status_code
```

```python
response.content
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
stat_table = soup.find_all('table', class_='tablehead')
```

```python
len(stat_table)
```

```python
type(stat_table)
```

```python
stat_table
```

```python
stat_table = stat_table[0]
```

```python
type(stat_table)
```

```python
stat_table
```

```python
for row in stat_table.find_all('tr'):
    for cell in row.find_all('td'):
        print(cell.text)
```

```python
with open ('D:\Basketball_Stats.csv', 'w') as r:
    for row in stat_table.find_all('tr'):
        for cell in row.find_all('td'):
            r.write(cell.text.ljust(24))
        r.write('\n')
    
```
